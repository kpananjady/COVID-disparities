```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('COVID-19_Cases_and_Deaths_by_Race_Ethnicity.csv')
```

```python
df['Date updated'] =pd.to_datetime(df['Date updated'])
```

```python
df.columns
```

# Death rate by race

```python
#How exactly is the age-adjusted death rate calculated?
#check missing data 
```

```python
sns.lineplot(data=df, x="Date updated", y="Age adjusted death rate per 100k", hue="Race/ethnicity")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xticks(rotation=45)
```

```python
sns.lineplot(data=df, x="Date updated", y="Crude death rate per 100k", hue="Race/ethnicity")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xticks(rotation=45)
```

```python
df = pd.read_csv('COVID-19_Cases_and_Deaths_by_Age_Group.csv')
```

```python
df['DateUpdated'] = pd.to_datetime(df['DateUpdated'])
```

```python
sns.lineplot(data=df, x="DateUpdated", y="Total case rate", hue="AgeGroups")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xticks(rotation=45)
```

```python
df['AgeGroups'].unique()
```

```python
df[df['AgeGroups']=='70-79']
```

```python
df['case_rate_pct_change'] = df[df['AgeGroups']=='80 and older'].sort_values(by='DateUpdated')['Total case rate'].pct_change()
```

```python
df[df['AgeGroups']=='80 and older'].plot(x='DateUpdated', y='case_rate_pct_change')
```

```python
df[df['AgeGroups']=='80 and older'].plot(x='DateUpdated', y='Total case rate')
```
